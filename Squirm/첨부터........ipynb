```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython.display
```

```python
DATA_DIR = "./wav/train/고생했어2.wav"

# wav : 사운드 데이터
# sr : sampling rate (초당 몇개의 소리 데이터로 구성)
wav, sr = librosa.load(DATA_DIR,sr=16000)

print("초당 샘플 수 :",sr)
print("wav 파일 길이 :", wav.shape)
print("플레이 시간 :", wav.shape[0] / float(sr), "초")
```

```python
plt.figure(figsize=(100, 30))
plt.plot(wav)

plt.show()
```

```python
wav.shape
```

```python
y=[]
for i in range(len(wav)):
    y.append(round(wav[i],3))
```

```python
plt.figure(figsize=(100, 30))
plt.plot(y)

plt.show()
```

```python
t=[]
for i in range(len(y)):
    if y[i]<=0.03:
        y[i]=0
    t.append(y[i])
```

```python
average =[]
for i in range(len(t)):
    if t[i]!=0:
        average.append(t[i])
average = np.array(average)
average = np.mean(average)
average
```

```python
t=[]
for i in range(len(y)):
    if y[i]<=average-0.02:
        y[i]=0
    t.append(y[i])
```

```python
plt.figure(figsize=(100, 30))
plt.plot(t)

plt.show()
```

```python
re_time =[]
count = 0
# and len(re_time)%2==0
for i in range(len(t)):
    if t[i]==0:
        count+=1
    elif t[i]>0 and count>=450:
        re_time.append(i)
        count=0
    else:
        count=0
```

```python
len(re_time)
```

```python
re_time
```

```python

for i in range(len(re_time)):
    if i == len(re_time)-1:
        plt.figure(figsize=(50, 30))
        plt.plot(t[re_time[i]:])
        plt.show()
    else:
        plt.figure(figsize=(50, 30))
        plt.plot(t[re_time[i]:re_time[i+1]])
        plt.show()
```
